```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tickers = ['XLB', 'XLE', 'XLF', 'XLI', 'XLK', 'XLP', 'XLU', 'XLV', 'XLY']
```

```python
yfobj = yf.Tickers(tickers)
```

```python
df = yfobj.history(start = '2011-01-02', end = '2025-03-02', actions = False)
```

```python
df.tail()
```

```python
df.drop(['Open','Low', 'High', 'Volume'], inplace=True, axis=1 )
```

```python
df.head()
```

```python
df.droplevel(0, axis = 1)
```

```python
df = df.resample('w').last()
```

```python
df.head()
```

```python
df.ffill(inplace=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.plot()
plt.title('Equity Close')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
rets = df.pct_change().fillna(0)
rets.plot()
plt.title('Equity Weekly Return')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
import numpy as np
```

```python
vols = rets.rolling(52).std() * np.sqrt(52)
vols.plot(figsize = (12,6))
plt.title('Equity Volatility')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
