```python
tickers = ['SPY', 'TLT']
etf =  yf.Tickers(tickers)
data  = etf.history(start = '2002-09-01', end = '2025-03-04', actions = False)
```

```python
data.tail()
```

```python
data.head()
```

```python
data.drop(['Open', 'Low', 'High', 'Volume'], inplace = True, axis=1)
```

```python
data
```

```python
data = data.droplevel(0, axis=1)
data
```

```python
data.plot()
```

```python
rets = data.pct_change().fillna(0)
rets.head()
```

```python
rets.plot()
plt.title('Historical Returns')
plt.xlabel('Date')
plt.ylabel('Returns')
```

```python
import numpy as np
```

```python
total_vols  = rets.std() * np.sqrt(252)
total_vols
```

```python
vols  = rets.rolling(window=252).std() * np.sqrt(252)
vols
```

```python
vols.plot()
plt.title('Historical Volatility')
plt.xlabel('Date')
plt.ylabel('Volatility')
```

```python
spy_rets = rets['SPY']
tlt_rets = rets['TLT']
total_corr = spy_rets.corr(tlt_rets)
total_corr

corr = spy_rets.rolling(window=252).corr(tlt_rets)
corr
```

```python
total_corr
```

```python
corr
```

```python
corr.plot()
plt.title('Historical Correlation')
plt.xlabel('Date')
plt.ylabel('Correlation')
```

```python
tickers = ['XLB', 'XLE', 'XLF', 'XLI', 'XLK', 'XLP', 'XLU', 'XLV', 'XLY']
```

```python
yfobj = yf.Tickers(tickers)
```

```python
df = yfobj.history(start = '2011-01-01', end = '2022-10-14', actions = False )
```

```python
df.tail()
```

```python
df.drop(['Open','Low', 'High', 'Volume'], inplace=True, axis=1 )
```

```python
df.ffill(inplace=True)
```

```python
df
```

```python
df.droplevel(0, axis=1)
df = df.resample('w').last()
```

```python
df
```

```python
df.plot()
plt.title('Historical Prices')
plt.xlabel('Date')
plt.ylabel('Price')
```

```python
rets = df.pct_change().fillna(0)
rets.plot(figsize=  (12,6))
plt.title('Historical Returns')
plt.xlabel('Date')
plt.ylabel('Returns')
```

```python
vols = rets.rolling(52).std() * np.sqrt(52)
vols.plot(figsize=  (12,6))
plt.title('Historical Volatility')
plt.xlabel('Date')
plt.ylabel('Volatility')
```

```python
corr = rets.corr()
corr
```
